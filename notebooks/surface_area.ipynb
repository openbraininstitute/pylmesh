```python
import pylmesh
import os 
```

```python
# The surface mesh 
mesh_path = "../data/meshes/neuron.obj"
```

## Computing the surface area using the Mesh 

```python
# Check if file exists
if not os.path.exists(mesh_path):
    print(f"Error: File not found: {mesh_path}")
else:
    print(f"File exists: {mesh_path}")
    print(f"File size: {os.path.getsize(mesh_path) / 1024:.2f} KB")

    try:
        mesh = pylmesh.load_mesh(mesh_path)
        print(f"\n✓ Loaded successfully!")
        print(f"Vertices: {mesh.vertex_count()}")
        print(f"Faces: {mesh.face_count()}")
    except Exception as e:
        print(f"\n✗ Failed to load: {e}")
        print("\nPossible issues:")
        print("- File may be corrupted")
        print("- Format may not be supported")
        print("- File may contain non-triangle geometry")

surface_area = mesh.surface_area()
print(f"\nSurface Area: {surface_area:.4f} units²")
pylmesh.save_mesh("../data/meshes/neuron_mesh.obj", mesh)
```

## Computing the surface area using the QuantizedMesh 

```python
# Check if file exists
if not os.path.exists(mesh_path):
    print(f"Error: File not found: {mesh_path}")
else:
    print(f"File exists: {mesh_path}")
    print(f"File size: {os.path.getsize(mesh_path) / 1024:.2f} KB")

    try:
        qmesh = pylmesh.load_quantized_mesh(mesh_path)
        print(f"\n✓ Loaded successfully!")
        print(f"Vertices: {qmesh.vertex_count()}")
        print(f"Faces: {qmesh.face_count()}")
    except Exception as e:
        print(f"\n✗ Failed to load: {e}")
        print("\nPossible issues:")
        print("- File may be corrupted")
        print("- Format may not be supported")
        print("- File may contain non-triangle geometry")

surface_area = qmesh.surface_area()
print(f"\nSurface Area: {surface_area:.4f} units²")

pylmesh.save_quantized_mesh("../data/meshes/neuron_qmesh.obj", qmesh)
```

```python
for i in [8, 10, 16, 21]:
    qmesh = pylmesh.load_quantized_mesh(mesh_path, bits_per_axis=i)
    print(f"\nQuantization Bits: {i}")
    print(f"Vertices: {qmesh.vertex_count()}")
    print(f"Faces: {qmesh.face_count()}")
    surface_area = qmesh.surface_area()
    print(f"Surface Area: {surface_area:.4f} units²")
```
