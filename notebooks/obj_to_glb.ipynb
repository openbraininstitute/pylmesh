# OBJ to GLB Conversion with pylmesh

This notebook demonstrates how to load OBJ mesh files and export them as GLB (binary GLTF) format.

## Import Library

```python
import pylmesh
import os

print(f"pylmesh version: {pylmesh.__version__}")
```

## Load OBJ File

```python
# Load an OBJ file
obj_file = "/scratch/quantization/12345.obj"  # Change to your OBJ file path

mesh = pylmesh.load_mesh(obj_file)

print(f"Loaded: {obj_file}")
print(f"Vertices: {mesh.vertex_count()}")
print(f"Faces: {mesh.face_count()}")
```

## Inspect Mesh Data

```python
# Show first few vertices
print("First 5 vertices:")
for i, v in enumerate(mesh.vertices[:5]):
    print(f"  v{i}: ({v.x:.3f}, {v.y:.3f}, {v.z:.3f})")

# Show first few faces
print(f"\nFirst 3 faces:")
for i, f in enumerate(mesh.faces[:3]):
    print(f"  f{i}: {f.indices}")
```

## Export to GLB

```python
# Export to GLB format
glb_file = "/scratch/quantization/output.glb"

pylmesh.save_mesh(glb_file, mesh)

print(f"Exported to: {glb_file}")
print(f"File size: {os.path.getsize(glb_file) / 1024:.2f} KB")
```

## Verify GLB Export

```python
# Load the exported GLB to verify
mesh_glb = pylmesh.load_mesh(glb_file)

print(f"Verified GLB file:")
print(f"Vertices: {mesh_glb.vertex_count()}")
print(f"Faces: {mesh_glb.face_count()}")
print(f"\nConversion successful! ✓")
```

## Batch Conversion

Convert multiple OBJ files to GLB:

```python
import glob

# Find all OBJ files in current directory
obj_files = glob.glob("*.obj")

print(f"Found {len(obj_files)} OBJ files\n")

for obj_file in obj_files:
    try:
        # Load OBJ
        mesh = pylmesh.load_mesh(obj_file)
        
        # Generate GLB filename
        glb_file = obj_file.replace(".obj", ".glb")
        
        # Export to GLB
        pylmesh.save_mesh(glb_file, mesh)
        
        print(f"✓ {obj_file} -> {glb_file} ({mesh.vertex_count()}v, {mesh.face_count()}f)")
    except Exception as e:
        print(f"✗ {obj_file}: {e}")

print("\nBatch conversion complete!")
```

## Export to Other Formats

pylmesh supports multiple formats:

```python
# Export to different formats
formats = [
    ("output.stl", "STL"),
    ("output.ply", "PLY"),
    ("output.off", "OFF"),
    ("output.gltf", "GLTF"),
    ("output.glb", "GLB")
]

for filename, format_name in formats:
    pylmesh.save_mesh(filename, mesh)
    size = os.path.getsize(filename) / 1024
    print(f"✓ {format_name:6} -> {filename:15} ({size:.2f} KB)")
```
