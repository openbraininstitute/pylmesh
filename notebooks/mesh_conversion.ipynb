# Mesh Format Conversion with pylmesh

Load and convert mesh files between different formats (OBJ, STL, PLY, OFF, GLTF, GLB).

## Import Library

```python
import pylmesh
import os

print(f"pylmesh version: {pylmesh.__version__}")
print("Supported formats: OBJ, STL, PLY, OFF, GLTF, GLB")
```

## Load Mesh File

Supports: .obj, .stl, .ply, .off, .gltf, .glb

```python
# Change to your mesh file path
input_file = "/scratch/quantization/12345.glb"

# Check if file exists
if not os.path.exists(input_file):
    print(f"Error: File not found: {input_file}")
else:
    print(f"File exists: {input_file}")
    print(f"File size: {os.path.getsize(input_file) / 1024:.2f} KB")
    
    try:
        mesh = pylmesh.load_mesh(input_file)
        print(f"\n✓ Loaded successfully!")
        print(f"Vertices: {mesh.vertex_count()}")
        print(f"Faces: {mesh.face_count()}")
    except Exception as e:
        print(f"\n✗ Failed to load: {e}")
        print("\nPossible issues:")
        print("- File may be corrupted")
        print("- Format may not be supported")
        print("- File may contain non-triangle geometry")
```

## Inspect Mesh Data

```python
if 'mesh' in locals():
    # Show first few vertices
    print("First 5 vertices:")
    for i, v in enumerate(mesh.vertices[:5]):
        print(f"  v{i}: ({v.x:.3f}, {v.y:.3f}, {v.z:.3f})")

    # Show first few faces
    print(f"\nFirst 3 faces:")
    for i, f in enumerate(mesh.faces[:3]):
        print(f"  f{i}: {f.indices}")
else:
    print("No mesh loaded. Run the previous cell first.")
```

## Export to Different Format

```python
if 'mesh' in locals():
    # Export to GLB
    output_file = "output.glb"
    
    try:
        pylmesh.save_mesh(output_file, mesh)
        print(f"✓ Exported to: {output_file}")
        print(f"File size: {os.path.getsize(output_file) / 1024:.2f} KB")
    except Exception as e:
        print(f"✗ Export failed: {e}")
else:
    print("No mesh loaded. Run the load cell first.")
```

## Export to All Formats

```python
if 'mesh' in locals():
    formats = [
        ("output.obj", "OBJ"),
        ("output.stl", "STL"),
        ("output.ply", "PLY"),
        ("output.off", "OFF"),
        ("output.gltf", "GLTF"),
        ("output.glb", "GLB")
    ]

    print("Exporting to all formats:\n")
    for filename, format_name in formats:
        try:
            pylmesh.save_mesh(filename, mesh)
            size = os.path.getsize(filename) / 1024
            print(f"✓ {format_name:6} -> {filename:15} ({size:.2f} KB)")
        except Exception as e:
            print(f"✗ {format_name:6} -> Failed: {e}")
else:
    print("No mesh loaded. Run the load cell first.")
```

## Batch Conversion

Convert multiple files in a directory:

```python
import glob

# Find all mesh files
input_dir = "/scratch/quantization/"  # Change to your directory
output_format = ".glb"  # Change to desired output format

# Find files with common mesh extensions
patterns = ["*.obj", "*.stl", "*.ply", "*.off", "*.gltf", "*.glb"]
files = []
for pattern in patterns:
    files.extend(glob.glob(os.path.join(input_dir, pattern)))

print(f"Found {len(files)} mesh files in {input_dir}\n")

for input_file in files[:5]:  # Limit to first 5 files
    try:
        mesh = pylmesh.load_mesh(input_file)
        
        # Generate output filename
        base = os.path.splitext(os.path.basename(input_file))[0]
        output_file = base + output_format
        
        pylmesh.save_mesh(output_file, mesh)
        
        print(f"✓ {os.path.basename(input_file):20} -> {output_file:20} ({mesh.vertex_count()}v, {mesh.face_count()}f)")
    except Exception as e:
        print(f"✗ {os.path.basename(input_file):20} -> Failed: {e}")

print("\nBatch conversion complete!")
```

## Create Test Mesh

If you don't have a mesh file, create a simple test mesh:

```python
# Create a simple triangle mesh
mesh = pylmesh.Mesh()

# Add 3 vertices
v1 = pylmesh.Vertex()
v1.x, v1.y, v1.z = 0.0, 0.0, 0.0

v2 = pylmesh.Vertex()
v2.x, v2.y, v2.z = 1.0, 0.0, 0.0

v3 = pylmesh.Vertex()
v3.x, v3.y, v3.z = 0.5, 1.0, 0.0

mesh.vertices = [v1, v2, v3]

# Add 1 face
face = pylmesh.Face()
face.indices = [0, 1, 2]
mesh.faces = [face]

print(f"Created test mesh: {mesh.vertex_count()}v, {mesh.face_count()}f")

# Save as GLB
pylmesh.save_mesh("test_triangle.glb", mesh)
print("Saved as: test_triangle.glb")
```
